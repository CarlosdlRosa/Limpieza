## Goal Analysis

## Second Halves Analysis

### Which team has suffered the most comebacks?

```python
# To answer the question above, we'll first create a dictionary 
# In this dictionary, each team's name is a key and the initial value is 0 comebacks
teams = df['HomeTeam'].unique()
comebacks_suffered = {team: 0 for team in teams}

# We iterate through each row of the DataFrame, which represents each match of the season, to count comebacks
# The .iterrows() method is particularly useful for this task, especially when combined with a 'for' loop
for index, row in df.iterrows():
    # If the home team loses after being ahead at halftime (HTR),
    # we count it as a comeback suffered by the home team
    if row['HTR'] == 'H' and row['FTR'] == 'A':  
        comebacks_suffered[row['HomeTeam']] += 1
    # We then apply the same logic for the away team
    elif row['HTR'] == 'A' and row['FTR'] == 'H': 
        comebacks_suffered[row['AwayTeam']] += 1

comebacks_suffered
```

### Which team has completed the most comebacks?

```python
# To answer this question, our code will be similar to the one used for the previous question.
# We need to create a dictionary with teams as keys and 0 as the initial value
teams = df['HomeTeam'].unique()
comebacks_completed = {team: 0 for team in teams}

# Then, we iterate through the matches to count the completed comebacks for each team
for index, row in df.iterrows():
    if row['HTR'] == 'A' and row['FTR'] == 'H':  
        comebacks_completed[row['HomeTeam']] += 1
    elif row['HTR'] == 'H' and row['FTR'] == 'A': 
        comebacks_completed[row['AwayTeam']] += 1

comebacks_completed
```

### Who has obtained the most wins in the second halves?


```python
# We can answer this question using a similar methodology to the one we used for the last two questions.
# We create a new dictionary that we will populate by iterating through the rows of our data
teams = df['HomeTeam'].unique()
second_half_wins = {team: 0 for team in teams}


# This time, we'll iterate through each match to detect when teams achieved the victory specifically in the 2nd half 
for index, row in df.iterrows():
    # If the home team was not ahead at halftime but won by the end, we count this as a second half win for the home team
    if row['HTR'] != 'H' and row['FTR'] == 'H':  
        second_half_wins[row['HomeTeam']] += 1
    # We apply the same logic for the away team
    elif row['HTR'] != 'A' and row['FTR'] == 'A': 
        second_half_wins[row['AwayTeam']] += 1

second_half_wins
```

### Which team earned the most points in the second halves?

```python
# This question will require a procedure similar to the previous ones to be answered
# We create a dictionary to track points earned by each team in the second halves
teams = df['HomeTeam'].unique()
second_half_points = {team: 0 for team in teams}

# This time, though, the iteration through the matches is slightly more complex
# Basically, we compare the points a team earned by half-time to the points they finally received
# If the final points are higher than the half-time points, the positive difference is added to the team’s count
for index, row in df.iterrows():
    if row['HTR'] == 'D' and row['FTR'] == 'H':  
        second_half_points[row['HomeTeam']] += 2
    elif row['HTR'] == 'D' and row['FTR'] == 'A':
        second_half_points[row['AwayTeam']] += 2
    elif row['HTR'] == 'H' and row['FTR'] == 'D': 
        second_half_points[row['AwayTeam']] += 1
    elif row['HTR'] == 'H' and row['FTR'] == 'A': 
        second_half_points[row['AwayTeam']] += 3
    elif row['HTR'] == 'A' and row['FTR'] == 'D': 
        second_half_points[row['HomeTeam']] += 1
    elif row['HTR'] == 'A' and row['FTR'] == 'H': 
        second_half_points[row['HomeTeam']] += 3

second_half_points
```

```python
# For a more comprehensive read of our data exploration outcome, let's sort the dictionary by the points in descending order
sorted_2nd_half_points = sorted(second_half_points.items(), key=lambda x: x[1], reverse=True)

sorted_2nd_half_points
```

```python
# The sorted() method always returns a list of tuples, so we convert this list back into a dictionary
sorted_2nd_half_points_dict = {team: points for team, points in sorted_2nd_half_points} 
sorted_2nd_half_points_dict
```

```python
# However, let's present the results of our exploration in a more appealing way 
# To do this, we create a plot with Seaborn where one axis represents the keys and the other axis represents the values of the dictionary
sns.barplot(x=sorted_2nd_half_points_dict.keys(), y=sorted_2nd_half_points_dict.values())
plt.xticks(rotation=60) # Used to rotate the x-axis labels
plt.title("Points Obtained in Second Halfs by Team")
```

## What do the final league standings look like?

### How many points did each team earn by the end of the league?

```python
# To find out how many points each team earned by the end of the league, 
# we need to create a dictionary with the names of all the teams again
teams = df['HomeTeam'].unique()
final_standings = {team: 0 for team in teams}

# This 'for' loop iterates through all the matches and, in each case, adds 3 points to the winner of the match.
for index, row in df.iterrows():
    if row['FTR'] == 'H':  
        final_standings[row['HomeTeam']] += 3
    elif row['FTR'] == 'A': 
        final_standings[row['AwayTeam']] += 3
    # If the match ends in a draw, the loop adds 1 point to both the home team and the away team
    elif row['FTR'] == 'D':
        final_standings[row['HomeTeam']] += 1
        final_standings[row['AwayTeam']] += 1

# As the final step, we sort the teams in descending order of points to simulate the league table
sorted_standings = sorted(final_standings.items(), key= lambda x: x[1], reverse=True)

sorted_standings
```

### Reconstruction of the final league standings with goals scored and conceded

```python
# We have already generated some data to include in the final league standings
# This is the case with the already sorted final standings, stored in the variable sorted_standings
# However, it's a tuple, so we'll need to create a DataFrame using the .DataFrame() method
# We also create two columns, 'Team' and 'Points', to split each tuple into these columns.
df_table = pd.DataFrame(data=sorted_standings, columns=['Team', 'Points'])

df_table
```

```python
# Now it's time to count the goals scored and conceded
# We will simultaneously create a dictionary for goals scored and another for goals conceded
goals_scored = {team: 0 for team in teams}
goals_conceded = {team: 0 for team in teams}


# By using this loop, we ensure that we count the goals scored and conceded by both the Home and Away teams for all the matches in the league
for index, row in df.iterrows():
    goals_scored[row['HomeTeam']] += row['FTHG']    # Add goals scored by the Home team
    goals_scored[row['AwayTeam']] += row['FTAG']    # Add goals scored by the Away team
    goals_conceded[row['HomeTeam']] += row['FTAG']  # Add goals conceded by the Home team
    goals_conceded[row['AwayTeam']] += row['FTHG']  # Add goals conceded by the Away team


# The .map() method maps the values by looking up the specified column in the DataFrame and finding them in the dictionary keys
# For each team name in df_table["Team"], .map() looks up the corresponding value in the goals_scored and goals_conceded dictionaries
# and assigns these values to the new columns "Goals scored" and "Goals conceded"
df_table["Goals scored"] = df_table["Team"].map(goals_scored)
df_table["Goals conceded"] = df_table["Team"].map(goals_conceded)

df_table
```

```python
# Let's create the 'Goal difference' column by calculating the difference between goals scored and goals conceded
df_table['Goal difference'] = df_table['Goals scored'] - df_table['Goals conceded']

df_table
```

```python
# We will have completed the necessary data for the final standings once we add the total wins, draws and losses for each team
# To do that, let's create simultaneously 3 dictionaries, one for each concept
wins_per_team = {team: 0 for team in teams}
draws_per_team = {team: 0 for team in teams}
losses_per_team = {team: 0 for team in teams}

# By using this loop, we'll count the number of wins, draws, and losses for each team
for index, row in df.iterrows():
    if row['FTR'] == 'H':
        wins_per_team[row['HomeTeam']] += 1
        losses_per_team[row['AwayTeam']] +=1
    elif row['FTR'] == 'A': 
        wins_per_team[row['AwayTeam']] += 1
        losses_per_team[row['HomeTeam']] +=1
    elif row['FTR'] == 'D':
        draws_per_team[row['HomeTeam']] += 1
        draws_per_team[row['AwayTeam']] += 1

# Let's map the wins, draws and losses data to the DataFrame using the same procedure we used for goals scored and conceded
df_table['Wins'] = df_table['Team'].map(wins_per_team)
df_table['Draws'] = df_table['Team'].map(draws_per_team)
df_table['Losses'] = df_table['Team'].map(losses_per_team)

df_table
```

```python
# To reorder the table columns in a more meaningful way, let's create a list with all the column names in the desired order
columns_ordered = ['Team', 'Points', 'Wins', 'Draws', 'Losses', 'Goals scored', 'Goals conceded', 'Goal difference']

# Next, we use the '.reindex()' method to apply the specified column order to the DataFrame
df_table = df_table.reindex(columns= columns_ordered)

df_table
```
